```python
from google.genai import types # TIPOS DE INFORMA√á√ÉO QUE A SDK SUPORTA

chat_config = types.GenerateContentConfig(system_instruction='Voc√™ √© um assistente pessoal e sempre responde de forma sucinta')
chat = client.chats.create(model=modelo,config=chat_config) # INSTANCIANDO O CHAT COM A CONFIGURA√á√ÉO
```

CRIANDO UM CHATBOT COM JANELA PERSISTENTE USANDO LOOP

```python
prompt = 'inicio'

while prompt != 'fim':
    prompt = input('Esperando prompt...: ')
    print('Prompt: ',prompt,'\n') # √â OBRIGAT√ìRIO TER O \n PARA QUE O CHATBOT FUNCIONE E ENTENDA QUE O COMANDO TERMINOU
    resposta = chat.send_message(prompt)
    print('Resposta: ',resposta.text,'\n','Para sair, digite fim','\n')
    
```
