#### Mtodo 1 - Adicionar uma coluna ao Dataframe do Pandas com um valor padro.
**Formas de fazer** <br>
1. Declarando uma lista como uma nova coluna <br>
2. Utilizando loc (Voc pode usar '.loc' com ':' para adicionar um valor <br> especificado para todas as linhas.) <br>
3. Utilizando .assign() <br>
4. Utilizando .insert() <br>
#####  Paramentros para insert
<li>loc: loc  um inteiro que  a localizao de uma coluna onde queremos inserir uma nova coluna. Isso mudar a coluna existente naquela posio para a direita.<br>
<li>column: uma string que  o nome de uma coluna a ser inserida.<br>
<li>value: valor  simplesmente o valor a ser inserido. Pode ser um inteiro, uma string, um float ou mesmo uma srie/lista de valores. Fornecer apenas um valor definir o mesmo valor para todas as linhas.<br><br>
**allow_duplicates**:  um valor booleano que verifica se uma coluna com o mesmo nome j existe ou no.

#### Mtodo 2 - Adicionar uma coluna ao Dataframe do Pandas com valores diferentes

2. Utilizando loc (Voc pode usar '.loc' com ':' para adicionar um valor especificado para todas as linhas.)
3. Utilizando .assign()
4. Utilizando .insert()
#####  Paramentros para insert
-loc: loc  um inteiro que  a localizao de uma coluna onde queremos inserir uma nova coluna. Isso mudar a coluna existente naquela posio para a direita.<br>
-column: uma string que  o nome de uma coluna a ser inserida.<br>
-value: valor  simplesmente o valor a ser inserido. Pode ser um inteiro, uma string, um float ou mesmo uma srie/lista de valores. Fornecer apenas um valor definir o mesmo valor para todas as linhas.<br>
-**allow_duplicates**:  um valor booleano que verifica se uma coluna com o mesmo nome j existe ou no.


```python
# 1. Declarando uma lista como uma nova coluna
df['Final_grade'] = ['A','D','B','A','A','C']
df
```

#### Removendo registros duplicados
[Mtodo drop_duplicates](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop_duplicates.html#pandas-dataframe-drop-duplicates) Retorne DataFrame com linhas duplicadas removidas.

```python
# Criando o dataset

frame_data = {'nome'  : ['Berry', 'Kara', 'Clark', 'Berry'], 
              'idade' : [28, 32, 52, 28], 
              'funo': ['Office Boy', 'Enfermeira', 'Ascensorista', 'Office Boy']}

df = pd.DataFrame(frame_data)
df
```

```python
# Verificado se existem registros duplicados
df.duplicated()
```

#### Ateno: por padro o mtodo drop_duplicates retorna um dataframe temporrio em memria, ento se quiser gerar um novo dataframe, faa uma atribuio para uma nova varivel.
#### Caso voc queira atualizar o dataframe atual, voc deve utilizar o parmetro inplace. (Como se fosse um commit)

```python
# Se for positivo, podemos remover os duplicados utilizando o metodo
df_no_duplicates = df.drop_duplicates()
df_no_duplicates
```

```python
# O dataframe original continua igual
df
```

```python
# Para alterar diretamente no dataframe
df.drop_duplicates(inplace=True)
```

```python
df
```
