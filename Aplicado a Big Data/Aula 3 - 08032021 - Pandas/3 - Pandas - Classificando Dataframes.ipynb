### Classificação

#### Classificar valores por linha/coluna
[Método sort_values](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sort_values.html#pandas-dataframe-sort-values) Classifique pelos valores ao longo de cada eixo.

Importante que os dados sejam do mesmo tipo

#### Classificando por coluna

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'col1': ['A', 'A', 'B', np.nan, 'D', 'C'],
    'col2': [2, 1, 9, 8, 7, 4],
    'col3': [0, 1, 9, 4, 2, 3],
    'col4': ['a', 'B', 'c', 'D', 'e', 'F']
})

df
```

```python
df.sort_values(by='col1')
```

#### Classificando por múltiplas colunas

```python
print(df)
df.sort_values(by=['col1', 'col2'])
```

#### Classificando de forma descendente

```python
print(df)
df.sort_values(by='col1', ascending=False)
```

#### Classificando colocando os NaNs para aparecer primeiro

```python
df.sort_values(by='col1', ascending=False, na_position='first')
```

#### Classificando linhas

```python
import pandas as pd

df = pd.DataFrame({0: [7, 2], 1: [3, 5], 2: [4, 8]})
df
```

```python
df.sort_values(by=0,axis=1)
```

#### Series não precisa de "by"

```python
df['col2'].sort_values()
```
