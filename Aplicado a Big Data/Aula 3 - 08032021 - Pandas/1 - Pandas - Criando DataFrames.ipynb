#### Bonus - Criando dataframes a partir de um dicionário de arrays NumPy
<li>Um datraframe Pandas é uma estrutura de dados de 2 dimensões, comio uma matriz de 2 dimensões, ou uma tabela com linhas e colunas.
<li><b>Importar a biblioteca pandas -></b> import pandas</b>
<li><b>Criar a estrutura de dados -></b> Matriz, dicionário, etc... 
<li><b>Inserir a estrutura de dados no método <i>DataFrame</i></b> -> DataFrame(data)

```python
# imports
import numpy as np
import pandas as pd

# Utilizando dicionários de arrays numpy

data = {'order':np.arange(start=0 , stop = 120 , step = 1),
        'temp':np.linspace(start=0, stop=120, num=120)}
        #'temp':np.linspace(start=-10, stop=150, num=120)}

# Criando dataframe com Ndarrays. Inserindo no dataframe
df = pd.DataFrame(data)

print(df)
```

```python
print(df.values)
```

#### Bonus 2.1 - Criando novas colunas em um dataframe

```python
# Criando dataframe de teste
import pandas as pd

students = [('Jack', 34, 'Sydney' , 'Australia'),
            ('Joshua', 30, 'Delhi' , 'India' ),
            ('Tom', 31, 'Mumbai' , 'India' ),
            ('Alicia', 32, 'Bangalore' , 'India' ),
            ('John', 16, 'New York' , 'US'),
            ('Osama', 17, 'las vegas' , 'US')]

df = pd.DataFrame(students, 
                columns = ['Name' , 'Age', 'City' , 'Country'], 
                index=['a', 'b', 'c' , 'd' , 'e' , 'f'])

print(df)
```

#### Descrevendo o dataframe
<li><b>describe</b>-> Descrição dos dados do dataframe
<ol>
    <b><li>count - </b>The number of not-empty values.
    <b><li>mean - </b>The average (mean) value.
    <b><li>std - </b>The standard deviation.
    <b><li>min - </b>the minimum value.
    <b><li>25% - </b>The 25% percentile*.
    <b><li>50% - </b>The 50% percentile*.
    <b><li>75% - </b>The 75% percentile*.
    <b><li>max - </b>the maximum value.
</ol>

<b>*Percentil</b>
<br><li>Quantos dos valores são menores do que o percentil. 
<br><li>Um percentil é uma medida estatística e representa cada uma das partes de 100 partes iguais de um conjunto ordenado de valores.
<br>Ex: Por exemplo, se uma criança no peso se encontra no percentil 15, significa que em 100 crianças saudáveis com a mesma idade
<br>e gênero, 85 são mais pesadas e 15 são mais leves que essa criança.
<br><br>
<b>Desvio Padrão</b> (Em porcentagem)
<br><li>O Desvio Padrão consiste em uma medida do nível de dispersão, isto é, ele indica quão uniforme está um conjunto de dados. 
<br><li>Quanto maior o Desv. Padrão, o conjunto de dados está mais distante da média.
<br><li>É frequentemente utilizado para identificar padrões, anomalias ou outliers


```python

# Descreva o dataframe
print("Descrevendo o dataframe")
print(df.describe())
```

<b>describe(include='all') -> </b> Informa as colunas do dataset que possuem o tipo de dado informado no include.
<br>'all' é o padrão.

```python

print("Utilizando describe(include='all')")
print(df.describe(include='all')) # Informa as colunas do dataset que possuem o tipo de dado informado no include, conforme saída da função info
print("Inteiro 32")
print(df.describe(include='int32'))
print("Float 64")
print(df.describe(include='float64'))  
print()
```
