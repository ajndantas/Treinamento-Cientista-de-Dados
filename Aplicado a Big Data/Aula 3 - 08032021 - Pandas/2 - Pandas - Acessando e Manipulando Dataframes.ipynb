## Acessando Dataframes

#### O uso do loc[] (locate) é para índices (linhas) do dataframe
À frente, será visto que também pode-se usar a forma de acesso do numpy com ":"

#### Acessando por valor do indice, são retornadas todas as colunas da linha(índice) (Uma linha, uma dimensão) informado - Usando Pandas Series (Uma única dimensão)

```python
# Criando dataframe de teste
import pandas as pd

students = [('Jack  ', 34, 'Sydney' , 'Australia'),
            ('Joshua', 30, 'Delhi' , 'India' ),
            ('Tom', 31, 'Mumbai' , 'India' ),
            ('Alicia', 32, 'Bangalore' , 'India' ),
            ('John', 16, 'New York' , 'US'),
            ('Osama', 17, 'las vegas' , 'US')]

df = pd.DataFrame(students, 
                columns = ['Name' , 'Age', 'City' , 'Country'])

print(df)

df.shape
```

#### OPERADOR ISIN

```python

import pandas as pd

data1 = {
  "name": ["Sally", "Mary", "John"],
  "age": [50, 40, 30]
}

df1 = pd.DataFrame(data1)
print('Dataframe 1')
print(df1,'\n')

data2 = {
  "name": ["John", "Mary", "Sally"],
  "age": [30, 40, 50]
}

df2 = pd.DataFrame(data2)
print('Dataframe 2')
print(df2,'\n')

data3 = {
  "name": ["John"]
}

df3 = pd.DataFrame(data3)
print('Dataframe 3')
print(df3,'\n')

# SEMPRE A MELHOR MANEIRA DE FAZER A COMPARAÇÃO, POIS NAO IMPORTA A POSIÇÃO E NEM OS ÍNDICES DOS ELEMENTOS NO DATAFRAME
print(df3['name'].isin(df1['name']),'\n')
print(df1['name'].isin(df3['name']),'\n')
print(df1['name'].isin(["John"]),'\n')

print(df1.isin(df2),'\n')

print(all(list(df1.isin(df2))),'\n')

# USANDO SERIES, NÃO IMPORTA A ORDEM DOS REGISTROS
print(df2.isin([30,40,50]))
print(df2.isin([50,40,30]))

df2.loc[(df2.isin([50,40,30])['name'] == False).index]
```

#### Aplicar funções em linhas / colunas
[Método Apply](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html#pandas-dataframe-apply) Aplique uma função ao longo de um eixo do DataFrame.<br>
Por padrão, a função é aplicada em <b>TODAS</b> as colunas. Para aplicar em linhas, utilize o parâmetro axis = 1

```python
import pandas as pd


def calc_sum(x):
  return sum(x)

data = {
  "x": [50, 40, 30],
  "y": [300, 1112, 42]
}

df = pd.DataFrame(data)
print("df\n",df)

df.apply(calc_sum)
```

```python
import pandas as pd
  
data = {
  "x": [300, 1112, 42],
  "y" : [1,2,3]
}

df = pd.DataFrame(data)
print("df\n",df)

df['tamanho'] = df.astype(str).apply(lambda row : row.str.len().sum(), axis=1)
tamanho_maximo = df['tamanho'].max()
indice = df.loc[df['tamanho'] == tamanho_maximo].index[0]
print('\n',df)
print("Indice: ",indice)
print(df.loc[[1]].to_string(header=False,index=False))
```

## Manipulando Dataframes

### Adicionar colunas ao Dataframe do Pandas

### Método 1 - Adicionar uma coluna ao Dataframe do Pandas com um valor padrão.
**Formas de fazer** <br>
1. Declarando uma lista como uma nova coluna <br>
2. Utilizando loc (Você pode usar '.loc' com ':' para adicionar um valor <br> especificado para todas as linhas.). Será visto em Manipulando Dataframes <br>
3. Utilizando .assign() <br>
4. Utilizando .insert() <br>

```python
# Criando dataframe de teste

import pandas as pd

students = [('Jack', 34, 'Sydney' , 'Australia'),
            ('Joshua', 30, 'Delhi' , 'India' ),
            ('Tom', 31, 'Mumbai' , 'India' ),
            ('Alicia', 32, 'Bangalore' , 'India' ),
            ('John', 16, 'New York' , 'US'),
            ('Osama', 17, 'las vegas' , 'US')]

df = pd.DataFrame(students, 
                columns = ['Name' , 'Age', 'City' , 'Country'], 
                index=['a', 'b', 'c' , 'd' , 'e' , 'f'])
df
```

#### 1. Declarando uma lista como uma nova coluna

```python
df['School'] = 'Training'
df
```

### 2. Utilizando loc
#### Itens permitidos a atribuição
<li>Um elemento single, como um caracter ou um numero
<li>Uma lista ou array
<li>Um fatiamento de objetos
<li> Também se pode utilizar o fatiamento do Numpy



```python
df.loc[:,'Grade'] = 'A'
df
```

#### 3. Utilizando .assign()

```python
df = df.assign(Duration_months='24')
df
```

#### 4. Utilizando .insert()
**DataFrame.insert(loc, column, value, allow_duplicates=True)**
#####  Parametros para insert
<li>loc: loc é um inteiro que é a localização de uma coluna onde queremos inserir uma nova coluna (Começa em zero). Isso mudará a coluna existente naquela posição para a direita.<br>
<li>column: uma string que é o nome de uma coluna a ser inserida.<br>
<li>value: valor é simplesmente o valor a ser inserido. Pode ser um inteiro, uma string, um float ou mesmo uma série/lista de valores. Fornecer apenas um valor definirá o mesmo valor para todas as linhas.<br><br>

**allow_duplicates**: é um valor booleano que verifica se uma coluna com o mesmo nome já existe ou não.



```python
# Loc=1 é a posição que a coluna entrará
df.insert(loc=1, 
        column='Start_year', 
        value='2017',allow_duplicates=False)
df
```

### Método 1 - Inserindo uma linha por meio de um dataframe

A estrutura de dados do dataframe que representa a nova linha, precisa ser a mesma do dataframe ao qual se quer adicionar a nova linha.</br>
<br>Ex: Se o dataframe original é uma lista de tuplas, o dataframe que representa a nova linha, também deverá ser.

<b>[df = pd.concat([df, new_record], ignore_index=True)](https://pandas.pydata.org/docs/reference/api/pandas.concat.html)</b>
<br><br> O ignore_index, no caso de True, é para fazer a contagem a partir do último índice do dataframe

```python
nova_linha = [('Antonio',46,'Rio de Janeiro','Brasil')]

df_nova_linha = pd.DataFrame(nova_linha,columns = df.columns)

df = pd.concat([df,df_nova_linha],ignore_index=True)
df
```

### Método 2 - Inserindo uma nova linha por meio de uma lista - Pandas Series

```python
print('Tamanho do dataframe: ',len(df))

df.loc[len(df)] = ['Enzo',10,'Rio de Janeiro','Brasil']
print(df)

print("O último registro é do tipo: ",type(df.loc[df.index[-1]]))
```

### Inserindo uma linha no topo

df = pd.concat(<b>[new_record,df]</b>,ignore_index=True) </br>

```python
nova_linha = [('Liana',47,'Rio de Janeiro','Brasil')]
df_nova_linha = pd.DataFrame(nova_linha,columns = df.columns)

df = pd.concat([df_nova_linha,df],ignore_index=True)
df
```

## Duplicatas

#### Verificando duplicatas

```python
# Criando o dataset

import pandas as pd

frame_data = {'nome'  : ['Berry', 'Kara', 'Clark', 'Berry'], 
              'idade' : [28, 32, 52, 28], 
              'função': ['Office Boy', 'Enfermeira', 'Ascensorista', 'Office Boy']}

df = pd.DataFrame(frame_data)
df
```

```python
# Verificado se existem registros duplicados
df.duplicated()  
```

#### Removendo registros duplicados
[Método drop_duplicates](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop_duplicates.html#pandas-dataframe-drop-duplicates) Retorne DataFrame com linhas duplicadas removidas.

```python
print(df.drop_duplicates())
```

#### Atenção: por padrão o método drop_duplicates retorna um dataframe temporário em memória, então se quiser gerar um novo dataframe, faça uma atribuição para uma nova variável.
#### Caso você queira atualizar o dataframe atual, você deve utilizar o parâmetro inplace. (Como se fosse um commit)

```python
# Se for positivo, podemos remover os duplicados utilizando o metodo
df_no_duplicates = df.drop_duplicates()
df_no_duplicates
```

```python
# O dataframe original continua igual
df
```

```python
# Para alterar diretamente no dataframe
# Atenção a regra do inplace. 
# Escolhe remover a última ocorrência.

df.drop_duplicates(inplace=True)
```
