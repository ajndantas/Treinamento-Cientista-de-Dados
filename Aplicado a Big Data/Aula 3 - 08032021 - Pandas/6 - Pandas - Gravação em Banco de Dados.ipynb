##### Extraindo linha a linha - fetchone()

```python
cursor = connection.cursor()
employee_cursor = cursor.execute('SELECT * FROM employees')

umalinha = employee_cursor.fetchone()
print("Uma linha: ",umalinha)

outralinha = employee_cursor.fetchone()
print("Outra linha: ",outralinha)
```

#### SELECT COM WHERE   


```python
from conexao import conexao as con

connection = con(senha="hr", usuario="hr", ip="192.168.0.8", sid="oracle")

cursor = connection.cursor()
codigoemployee = '104'

employee_cursor = cursor.execute('SELECT * FROM employees WHERE employee_id = :employee_code',employee_code = codigoemployee)

umalinha = employee_cursor.fetchone()
print("Uma linha: ", umalinha)
```

#### SELECT COM WHERE IN

```python
from conexao import conexao as con

connection = con(senha="hr", usuario="hr", ip="192.168.0.8", sid="oracle")

cursor = connection.cursor()
codigosemployees = ('104','200','203') # TEM QUE UMA TUPLA

employee_cursor = cursor.execute(f'SELECT * FROM employees WHERE employee_id IN {codigosemployees}')

todasaslinhas = employee_cursor.fetchall()
print("Todas as linhas: ", todasaslinhas)
```
