```python
import timeit


def map_function(numbers):
    return list(map(lambda x: x * 2, numbers))


def for_loop_function(numbers):
    result = []
    for num in numbers:
        result.append(num * 2)
    return result


numbers = list(range(1000000))

map_time = timeit.timeit(lambda: map_function(numbers), number=10)
for_loop_time = timeit.timeit(lambda: for_loop_function(numbers), number=10)

print("Map time:", map_time)
print("For loop time:", for_loop_time)
```
