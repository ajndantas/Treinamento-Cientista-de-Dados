### Passagem por referência

##### A diferença é que está sendo passada uma variável


```python
'''
    Passagem de parâmetro por referência
'''

def changeme(mylist):
    print("Valores antes da alteração:",mylist)
    mylist[2] = 50
    print("Valores depois da alteração:",mylist)

mylist = [10,20,30]
changeme(mylist) # Passagem de parâmetro por referência. A diferença é que está 
                 # sendo passada uma variável

print("Valores de fora da função:",mylist) # A função já foi modificada. 
```

```python
def changeme(mylist):
    mylist = [1,2,3,4] # Fez uma nova atribuição para a varíavel
    print("Valores antes da alteração:",mylist)
    mylist[2] = 50
    print("Valores depois da alteração:",mylist)

mylist = [10,20,30]
changeme(mylist) # Passagem de parâmetro por referência

print("Valores de fora da função:",mylist) # mylist não está no escopo da função
```

### Passagem por chave

```python
'''
 Passagem de parâmetro por chave

  Quando você usa argumentos de palavra-chave em uma chamada de função, o chamador       identifica os    argumentos pelo nome do parâmetro. Que devem ter os mesmos nomes      dos que foram definidos na função.

  Isso permite colocar os arqgumentos fora de ordem, porque o intérprete  Python  pode   usar  as  palavras-chave  fornecidas  para  combinar  os valores com os parâmetros.
'''

def printinfo(nome, idade):
    print("Nome:",nome)
    print("Idade:",idade)

    return

printinfo(idade = 31, nome = "Saul Goodman") # A passagem por chave permite trocar a ordem.                                             
```

### Passagem por número variável de parâmetros

```python
'''
    Passagem com número variável de parâmetros.

    A variável que representará os n parâmetros recebe * antes
'''

def printinfo(arg1, *vartuple):
        print("A saída é: ")
        print(arg1)

        for var in vartuple: # Precisa de um for para a variável com número variável de parâmetros
            print(var)
        
        return

printinfo(10)
printinfo(70,60,50)
```

### Com valor padrão

```python
'''
    Função definida com valor padrão.

    Dessa forma, não precisamos, necessariamente, fornecer o valor para o argumento
    definido com valor padrão
'''

def printinfo(nome, idade = 35):
    print("Nome:",nome,"Idade",idade)

printinfo("Saul Goodman")

printinfo("John Watson",40)
```

### Uso do return

```python
'''
    Uso do return
'''

def soma(arg1, arg2):
    final = arg1 + arg2
    print("Dentro da função",final)
    return final

total = soma(10,20) # Aqui o print está atuando. O return não tem efeito algum

print("Fora da função",total) # Aqui o return está atuando
```

### Escopo de variáveis

```python
'''
    Escopo de variáveis
'''

total = 0 # Escopo global

def soma(arg1, arg2):
    total = arg1+arg2
    print("Dentro da função",total)
    
soma(10,20) 

print("Fora da função, global:",total)
```

```python
Money = 2000


def AddMoney():
    #global Money
    Money = Money + 1
    return Money

print(Money)

'''
     Dá erro a partir daqui, pois a variável Money não foi definida ainda na função,
     para corrigir, precisamos descomentar a linha com o global, dessa forma, estaremos considerando a declaração que foi feita em contexto global
'''
AddMoney()
print(Money)
```

```python
Money = 2000


def AddMoney():
    global Money
    Money = Money + 1

print(Money)
AddMoney()
print(Money)  # Uma vez considerada em contexto global, ela ganhou novo valor
```
