# Função Anônima ou Lambda

```python
'''
    Função anônima ou Lambda
'''


def soma(arg1, arg2): return arg1 + arg2


# Tem que imprimir dessa forma uma função lambda. Usando a varíável que recebe o seu valor
print("O valor da soma é:", soma(10, 20))
```

```python
# Loucura

def soma(a,b):
    print(a + b)

operacao = lambda arg1, arg2 : soma(arg1,arg2)
#operacao(10,15)

print("O valor da operação é:",operacao(10,20))
```
