# Módulo time

## time.asctime() 
#### Retorna uma  sequência  legível de  24  caracteres, como 'Ter 11 de dezembro, 18:07:14 2008'

```python
import time

tempo = time.asctime()
print("Tempo atual:",tempo)
```

## time.localtime() 
#### Retorna uma tupla de tempo (tupletime) referente ao tempo atual

```python
import time

tempo = time.localtime()
print("Tupla de tempo (tupletime) referente ao tempo atual:\n",tempo)
print("tm_isdt = 0 -> horário de verão não ativo \ntm_isdt = 1 -> horário de verão ativo")

print("Ano:",tempo[0])
print("Mes:",tempo.tm_mon)
print("Dia:",tempo.tm_mday)
```

## time.strptime() 
#### Python time method strptime() parses a string representing a time according to a format. The return value is a struct_time (tuple time)
#### Referência: https://www.tutorialspoint.com/python/time_strptime.htm

```python
import time

struct_time = time.strptime('01/04/2023 21:35:00', '%d/%m/%Y %H:%M:%S')
print (struct_time)
```

```python
# OBTENDO FINAIS DE SEMANA ENTRE DUAS DATAS
from datetime import date, datetime, timedelta
from typing import Dict

def obtem_finaisdesemana(data_inicio_mes_competencia: date, data_fim_mes_competencia: date) -> Dict:
    
    data = data_inicio_mes_competencia
    finaisdesemana = {}
    sabado = []
    domingo = []
    
    while data <= data_fim_mes_competencia:
        if data.weekday() == 5:            
            sabado.append(data.strftime('%Y-%m-%d'))
        elif data.weekday() == 6:
            domingo.append(data.strftime('%Y-%m-%d'))

        data = data + timedelta(days=1)
        
    finaisdesemana = {'sabado': sabado, 'domingo': domingo}
    
    return finaisdesemana

data_inicio_mes_competencia = datetime.strptime('15/04/2025','%d/%m/%Y').date()
data_fim_mes_competencia = datetime.strptime('15/05/2025','%d/%m/%Y').date()

print((data_fim_mes_competencia - data_inicio_mes_competencia).days)

print(obtem_finaisdesemana(data_inicio_mes_competencia,data_fim_mes_competencia))
```

# Módulo calendar

### month(year,mês) -> Exibe o calendário de um determinado mês e ano

```python
import calendar

calendario_mes = calendar.month(2023,4)
print(calendario_mes)
```

### calendar(year) -> Exibe o calendário para um determinado ano

```python
import calendar

calendario_ano = calendar.calendar(2023)
print(calendario_ano)
```
