### Yield
O Yield vai criar um generator no Python<br><br>
A função Yield pode fazer uma grande diferença no desempenho e eficiência do seu código, especialmente quando estamos lidando com grandes quantidades de informações.
<br><br>Ao invés de abrir o arquivo todo e ler de uma única vez, o Yield irá ler linha por linha do arquivo, sem trazer todas as informações de uma vez. (Lazy)
