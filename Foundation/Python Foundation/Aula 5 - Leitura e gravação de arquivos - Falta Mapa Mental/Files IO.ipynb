## Navegando por toda estrutura de diretórios
### Método os.walk(\<datapath\>)

#### Exemplo 1

```python
import os

# Armazene em uma variável o local onde estarão os arquivos que serão utilizados
data_path = 'G:\\Meu Drive\\Cursos e Treinamentos\\Cientista de Dados\\Treinamento Python\\Foundation\\Python Foundation\\Aula 5 - Leitura e gravação de arquivos - Falta Mapa Mental\\Minha Pasta'

def main():
    
    # PRIMEIRO DIRETÓRIO INFORMADO
    files = os.walk( data_path )

    for file in files:
        print(file[0])            
                

if __name__ == "__main__":
    main()
    
```

#### Exemplo 2

```python
import os

# Armazene em uma variável o local onde estarão os arquivos que serão utilizados
data_path = 'G:\\Meu Drive\\Cursos e Treinamentos\\Cientista de Dados\\Treinamento Python\\Foundation\\Python Foundation\\Aula 5 - Leitura e gravação de arquivos - Falta Mapa Mental\\Minha Pasta'

def main():
    
    # PRIMEIRO DIRETÓRIO INFORMADO
    files = os.walk( data_path )

    for file in files:
        dirpath = file[0]
        if len(file[2]) !=0:
            filenames = file[2]
            for file in filenames: 
                print(f'{dirpath}\\{file}')            
                

if __name__ == "__main__":
    main()
    
```

```python
# Lendo o conteúdo do arquivo
umArq = open("umArq.txt","r+")

str = umArq.read(8)
print("A leitura da string é:",str)

umArq.close()
```

```python
# Deslocando-se pelo arquivo
umArq = open("umArq.txt","r+")

str = umArq.read(-1)
    
print("A leituda da string é:",str)

print("A posição atual é:",umArq.tell())

umArq.seek(0,1) # O primeiro argumento informa a posição inicial da string, a partir 
                # de onde vamos contar, o segundo arqgumento, "1", informa que vamos 
                # começar a contar a partir da posição de onde paramos. Se fosse 0, 
                # iriámos voltar tudo desde o início.

str = umArq.read(8)
print("A nova string, a partir de onde paramos exibe:",str)

umArq.close()
```
