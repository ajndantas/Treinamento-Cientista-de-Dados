## Criando e acessando listas


```python

UmaLista = ['xwyz',123,5.15,'Dexter',92.5]
listinha = [456,'Corleone']
print(UmaLista)
print(listinha*2)
print(UmaLista + listinha)
```

### Acesso por ÍNDICE

```python
print("UmaLista[0]: " + UmaLista[0])
```

### Acesso por ÍNDICE : POSIÇÃO

```python
# Acesso posicional, considera o "1" como início, por isso é "3" em vez de "2", igual as strings

print("UmaLista[1:3]: ", UmaLista[1:3]) # Considera o "0", contando desde a posição "0", e não a partir do especificado.
print("UmaLista[2:]: ", UmaLista[2:]) # Considera o "0", contando a partir do especificado.
```

```python
lista1 = ['matemática','português',1992,2005]
print("Valor disponível on índice 2: ", lista1[2])
lista1[2] = 2001
print ("Novo valor disponível no índice 2: ",lista1[2])
```

## Excluindo valores da lista

```python
print(lista1)
del lista1[2]
print("Após excluir o valor no índice 2 da lista1: ",lista1)
```

## Incluindo e sobrescrevendo valores na lista passando a localização

```python
lista1[2:3] = [2020,'Enzo']
print(lista1)
```

## INDEX
The index() method returns the position at the first occurrence of the specified value.

```python
fruits = ['apple', 'banana', 'cherry']

x = fruits.index("cherry")
print (x)
```

## Índice negativo
Quando negativo, conta-se a partir da direita, e o índice inicial não é zero, é -1

```python

lingProg = ['Javascript','Java','Python']

print(lingProg[-2]) # 
```

## Ordenação, com critério


```python
# Utilizando critério de ordenação, key, a saída de uma função
# O critério de ordenação escolhido foi o tamanho da string ao reverso, ou seja, da maior para a menor.

def myFunc(e):
    return len(e)

print(lingProg)
lingProg.sort(key=myFunc,reverse=True)
print(lingProg)
```

## Ordenação, sem critério


```python
print(lingProg)
lingProg.sort()
print(lingProg)
```

## List Comprehension

## Acessando elementos de Sublista

```python
# Acesso lista dentro de lista
lista1 = ['matemática', 'português', 2020, 'Enzo', 'Liana',['Camila','Luciano']]
print(lista1[5])
print(lista1[5][0])  # Matriz de mais de uma dimensão
```

## Extendendo listas

```python
lista = [[1, 2], [3, 4], [5, 6]]

"""
    for sub in lista:
        for item in sub:
            print(item)
            
    O mesmo que o list comprehension abaixo
"""

flat = [item for sub in lista for item in sub]

print("Flat list:", flat)
```
