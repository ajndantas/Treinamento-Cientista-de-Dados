```python
print(9//2) # Divisão  de  piso (Floor  Division) - A  divisão
            # dos operandos em que o resultado é o quociente no qual  os  dígitos  após  o  ponto                  
            # decimal são removidos. Mas se um dos operandos for negativo, o  resultado  será                      
            # calculado,  ou  seja, arredondado para longe de zero (em direção ao infinito negativo):

print(9.0//2.0)
print(11//3)
print(11.0//3)
print(-11//3)
print(-11.0//3)
```

```python
'''
    OPERADORES IN e NOT IN
'''
from re import search

print (2 in [1,2,3,4])
print (5 not in [1,2,3,4])
print([1,2,3,4] in [1,2,3,4,[1,2,3,4]])
lista = ['a', 'b', 'c','felicidade']
print ('a' in lista)

encontrado = any(search(r'.*elicidade',elemento) for elemento in lista)
print('Encontrado' if encontrado else 'Não encontrado')



```
