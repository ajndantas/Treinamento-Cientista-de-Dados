### <b>SET</b>

##### Sets are used to store multiple items in a single variable.
##### Set is one of 4 built-in data types in Python used to store collections of data, the other 3 are List, Tuple, and Dictionary, all with different qualities and usage.
##### <b>A set is a collection which is unordered, unchangeable*, and unindexed.</b>
##### <b>* Note1: Set items are unchangeable, but you can remove items and add new items.</b>
##### <b>* Note2: Set items do not allow duplicate values.

```python
thisset = {"apple", "banana", "cherry"}
print(thisset)
```

##### <b>* Note2: Set items do not allow duplicate values.

```python
thisset = {"apple", "banana", "cherry", "apple"}

print(thisset)
```

##### Uma maneira para ordenar sets, é transformá-los em listas

```python
listaset = list(thisset)
listasetordenada = sorted(listaset)

print('Lista ordenada: ',listasetordenada)
```
