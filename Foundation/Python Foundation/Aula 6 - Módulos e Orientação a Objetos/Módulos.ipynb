# Módulos

### Importando o módulo
Um arquivo chamado modulos.py, neste caso, na mesma pasta do código principal

```python
import modulos # Importando o módulo modulos.py

modulos.print_func("Robocop")
```

```python
'''
    Função dir()
    
    A  lista  contém  os  nomes  de  todos  os  módulos,  variáveis e  funções
'''

import math
contem = dir(math)

print(contem) # Mostra todo o conteúdo de um módulo
```

```python
import fibonacci

contem = dir(fibonacci)

print(contem)
```

```python
'''
    A varíável sys.path

    O caminho de pesquisa do módulo é armazenado no módulo do sistema sys  como  a variável sys.path . 
    A  variável  sys.path  contém  o  diretório  atual, PYTHONPATH, e o padrão dependente da instalação. 
'''
import sys

print(sys.path)
```
