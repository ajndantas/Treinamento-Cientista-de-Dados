## Função match
<li>Busca a partir do índice zero
<li>O 'r' força que leia exatamente do jeito que foi passado 
<li><b>Objeto Match:</b> re.Match object; span=(0, 2), match='AV'. O span significa posição 0 até posição 2

```python
import re

frase = 'AV Analytics Vidhya AV'
result = re.match(r'AV',frase)

print(result.group())
                                        
```

```python
import re

result = result = re.match(r'Analytics','AV Analytics Vidhya AV')
# Analytics não está no índice zero

print(result)
```

## Função group
<li>Obtém o resultado do objeto match
<li><b>group(n)</b> n -> Ocorrência em que se obteve o resultado. "0" é a primeira 

```python
print(result.group())

print(result.group(0))# Buscando o parâmetro da primeira ocorrência
                      # se tentar com outro grupo, dará erro, pois o match busca a primeira ocorrência

```

```python
import re

result = re.match(r'AV','AV Analytics Vidhya AV')
print(result)

# Busca os índices de ocorrência no início e no fim do padrão, ao se encontrar,
# concatena as duas ocorrências e dá as posições iniciais das mesmas
print(result.start())
print(result.end())

```

## Função search
O search não busca somente na primeira posição, mas é apenas a primeira ocorrência

```python
import re

# O search não busca somente na primeira posição, mas é apenas a 
# primeira ocorrência

result = re.search(r'AV','AV Analytics Vidhya AV')
print(result)

print(result.group())
```

```python
# QUANDO NÃO ENCONTRA A STRING, O RETORNO É NONE
# O search não busca somente na primeira posição, mas é apenas a 
# primeira ocorrência
import re

result = re.search(r'AV','AV Analytics Vidhya AV')
print(result)
```

## Função findall
Retorna todas as correspondências não sobrepostas do padrão em string, como uma lista de strings ou tuplas. 

```python
import re

result = re.findall(r'AV','AV Analytics Vidhya AV')

print(result)
```

```python
from re import findall

query = "SELECT * FROM TABELA LIMIT 1234"
limit = int(findall(r'LIMIT (\d+)', query)[0])

print('limit: ',limit)
```

## Função split
Recorta a string em outras várias, excluíndo o termo passado, e os coloca em um array. 

```python
import re

result = re.split(r'i','Analytics Vidhya')

print(result)

resultx = re.split(r'i','Analytics Vidhya Life',maxsplit = 3) # Só faz uma divisão, o outro elemento, é o resto da string                                                         

print(resultx)
```

## Função sub
Faz replace na regex

```python
# Faz replace na regex

import re

result = re.sub(r'da Índia','do Mundo','AV é a maior comunidade de Analytics da Índia')

print(result)
```

```python
'''
    compile

    Uma maneira melhor de passar o padrão(regex) para a função
''' 

import re

pattern = re.compile('AV')

result = pattern.findall('AV Analytics Vidhya AV')
print(result)
```

```python
import re  

result=re.findall(r'.','AV is largest Analytics community of India') 

print(result)
```

```python
import re  

result=re.findall(r'\w','AV is largest Analytics community of India') 

print(result)
```

```python
import re  

result=re.findall(r'[a-zA-Z]{1}','AV is largest Analytics community of India') 

print(result)
```
