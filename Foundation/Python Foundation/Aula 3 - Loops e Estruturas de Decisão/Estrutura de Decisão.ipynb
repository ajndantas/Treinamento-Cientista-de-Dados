# IF

```python
num = 25
if num > 0:
    # O que estiver identado, faz parte da estrutura de decisão
    print(num, 'é um número positivo')
    
print("Esse texto sempre será exibido, não faz parte da estrutura de decisão")

num = -31
if num > 0:
    print(num, 'é um número positivo')
print("Esse texto sempre será exibido, não faz parte da estrutura de decisão")

if 2 > -3.5:
    print('Número maior')
```

# IF ELSE

```python
num = 52
if num >= 0:
    print("Positivo ou zero")
else:    
    print("Negativo")
```

# IF ELIF ELSE

```python
num = 78.9
if num > 0:
    print ("Número positivo")
elif num == 0:
    print ("Zero")
else:
    print("Número negativo")
```

# Condicional Aninhado

```python
# Condicional aninhado
num = 1000.0
if num >=0:
    if num == 0:
        print("Zero")
    else:
        print("Número positivo")
else:
    print("Número negativo")
```
