# FOR

```python
nums = [18,30,21,9,43,29,32,2,110]
soma = 0
for val in nums:
    soma = soma+val

print("O resultado da soma é: ",soma)
```

# FOR ELSE

```python
items = [0,1,5]
for i in items:
    print(i)
else:
    print("Não há mais itens")
```

## break

```python
student_name = 'Gohan'

registers = {'Goten':8.5,'Trunks':5.0, 'Majin Boo':9.0}

for student in registers: # student é preenchido com as chaves
    if student == student_name:
        print(registers[student])
        break
else: # Interessante essa posição para o else
     print('Não há mais resultados')
```

# WHILE

```python
n = 15
soma = 0
x = 1

while x<=n:
    soma += x
    x += 1

print(soma)
```

# WHILE ELSE

```python
count = 0

while count < 3:
    print("Dentro do while")
    count = count + 1
else:
    print("Dentro do else")
```

## continue

```python
i = "Uma frase"
n = 0
while i[n] != 'f':
    print("O valor de n é: ",n)
    n += 1
    continue    
print("O valor de i[n] é",i[n])
```
